```python
import numpy as np
import pandas as pd

# import matplotlib.pyplot as plt
```

```python
air_bnb = pd.read_csv('AB_NYC_2019.csv')
air_bnb.head()
```

```python
# How many neighborhood groups are available and which shows up the most?
avail = air_bnb['availability_365'] > 0 
neighbourhood_avail =air_bnb[avail]

neighbourhood_avail.groupby(['neighbourhood_group']).count()['availability_365']
```

```python
# Are private rooms the most popular in manhattan?

air_bnb['days_booked'] = 365 - air_bnb['availability_365']

air_bnb.groupby(['neighbourhood_group' , 'room_type']).mean(numeric_only=True).reset_index()[['neighbourhood_group','room_type','days_booked']].sort_values('days_booked', ascending=False)




# Yes private room are the most popular in Manhattan. 
```

```python
# Which hosts are the busiest and based on their reviews?

air_bnb.groupby(['host_name']).sum().sort_values('number_of_reviews', ascending=False).reset_index()[['host_name','number_of_reviews']]

#Michael David and john the busiest and based on their reviews
```

```python
#Which neighorhood group has the highest average price?
air_bnb.groupby(['neighbourhood_group']).mean(numeric_only=True).sort_values('price', ascending=False)

# Mannhattan have the highes avg. Price
```

```python
# Which neighbor hood group has the highest total price?
air_bnb.groupby(['neighbourhood_group']).sum(numeric_only=True).sort_values('price', ascending=False)

#Manhattan have the hightest Total Price
```

```python
#Which top 5 hosts have the highest total price?
air_bnb.groupby(['host_name']).sum(numeric_only=True).sort_values('price', ascending=False).reset_index()[['host_name','price']][:5]

# Sonder (NYC) have the highest total price: 82795
```

```python
# Who currently has no (zero) availability with a review count of 100 or more?
air_bnb.head()

 
host_names = []

def notAvailable(da):
    for index,i in da.iterrows():
        
        if i['number_of_reviews'] >= 100 and i['availability_365']==0:
            host_names.append(i['host_name'])
    return host_names

notAvailable(air_bnb)
host_names = set(host_names)
print(host_names)
```

```python
# What host has the highest total of prices and where are they located?

air_bnb.groupby(['host_name','neighbourhood_group']).sum(numeric_only=True).sort_values('price', ascending=False).reset_index()[0:1]

# Sonder(NYC) Located in Manhattan 
```

```python
# When did Danielle from Queens last receive a review?

dani = (air_bnb['host_name'] == 'Danielle') & (air_bnb['neighbourhood_group'] =='Queens')
air_bnb[dani].sort_values('last_review', ascending=False)[:1]

# Danielle from Queens last receive a review on 2019-07-08	
```
